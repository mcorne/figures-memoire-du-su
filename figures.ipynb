```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
q = 1  # Charge (en Coulombs)
a = 1  # Accélération (en m/s²)
c = 3e8  # Vitesse de la lumière (en m/s)
theta = np.linspace(0, 2 * np.pi, 500)  # Angles de 0 à 2π

# Valeurs fixes de r
r_values = [1, 2, 3]  # Différentes distances radiales (en m)
colors = ['blue', 'green', 'red']  # Couleurs pour les courbes

# Création du diagramme polaire
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})  # Taille normale pour le graphe

# Tracé des courbes pour chaque r
for r, color in zip(r_values, colors):
    flux = (q**2 * a**2 * np.sin(theta)**2) / (4 * np.pi * c**3 * r**2)  # Calcul du flux avec accélération
    ax.plot(theta, flux, label=f"$r = {r}$ m", color=color)

# Ajouter des cercles radiaux
max_flux = (q**2 * a**2) / (4 * np.pi * c**3 * min(r_values)**2)  # Flux max pour le plus petit r
radial_ticks = [0.25 * max_flux, 0.5 * max_flux, 0.75 * max_flux, max_flux]  # Niveaux de flux
ax.set_rticks(radial_ticks)  # Positions des cercles radiaux

# Doubler la taille des annotations des cercles radiaux
ax.set_yticklabels([f"{tick:.2e} W/m²" for tick in radial_ticks], fontsize=16)  # Taille augmentée

# Augmentation des tailles des angles et des légendes
ax.set_xticks([0, np.pi / 2, np.pi, 3 * np.pi / 2])  # Positions des angles principaux
ax.set_xticklabels(["$0°$", "$90°$", "$180°$", "$270°$"], fontsize=16)  # Taille augmentée pour les angles

# Ajuster les étiquettes de la légende
ax.legend(loc="upper right", fontsize=16)  # Taille augmentée pour les légendes

# Titre
# ax.set_title("Flux de Poynting proportionnel à $\\sin^2(\\theta)$ pour $r = 1, 2, 3$", va='bottom', fontsize=16)

# Affichage
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Vitesse relative en fraction de c (v/c)
v_over_c = np.linspace(0, 0.99, 500)  # De 0 à 0.99 c

# Facteur de Lorentz (gamma)
gamma = 1 / np.sqrt(1 - v_over_c**2)

# Tracé du facteur de Lorentz
plt.figure(figsize=(8, 6))
plt.plot(v_over_c, gamma, label=r"$\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$", color="blue")
# plt.axvline(0.9, color="red", linestyle="--", label=r"$v = 0.9c$")
plt.xlabel(r"Vitesse relative $v/c$", fontsize=14)
plt.ylabel(r"Facteur de Lorentz $\gamma$", fontsize=14)
plt.title("Variation du facteur de Lorentz en fonction de la vitesse", fontsize=16)
plt.grid(True, linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition des valeurs de beta typiques
beta_values = [0, 0.3, 0.6, 0.9, 0.99, 0.999]

# Définition des angles en degrés et conversion en radians
theta_degrees = np.linspace(0, 180, 100)  # 100 points entre 0° et 180°
theta_radians = np.radians(theta_degrees)

# Tracé des courbes pour différentes valeurs de beta
plt.figure(figsize=(8, 5))

for beta in beta_values:
    factor_values = (1 / (1 - beta * np.cos(theta_radians)))**5
    plt.plot(theta_degrees, factor_values, label=fr'$\beta = {beta}$')

# Labels et titre
plt.xlabel(r'$\theta$ (degrés)')
plt.ylabel(r'$1 / (1 - \beta \cos \theta)^5$')
# plt.title('Variation du facteur de correction avec $\theta$ pour différentes valeurs de $\beta$')
plt.yscale('log')  # Échelle logarithmique pour mieux voir la variation
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Affichage du graphique
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition des paramètres pour l'émissivité synchrotron
nu = np.logspace(8, 19, 300)  # Fréquence en Hz
p = 2.5  # Indice de distribution en énergie des électrons
nu_a = 1e10  # Fréquence d'auto-absorption (Hz)
nu_c = 1e18  # Fréquence de coupure (Hz)
alpha = (p - 1) / 2  # Indice spectral

# Définition des domaines de fréquence
nu_low = nu[nu < nu_a]  # Domaine auto-absorption
nu_mid = nu[(nu >= nu_a) & (nu < nu_c)]  # Domaine optiquement mince
nu_high = nu[nu >= nu_c]  # Domaine coupure

# Calcul des trois régimes séparément
epsilon_low = (nu_low / nu_a)**(5/2)  # Régime auto-absorption synchrotron
epsilon_mid = (nu_mid / nu_a)**(-alpha)  # Régime optiquement mince
epsilon_high = (nu_high / nu_a)**(-alpha) * np.exp(-nu_high / nu_c)  # Régime avec coupure exponentielle

# Tracé des courbes avec les styles demandés
plt.figure(figsize=(8, 6))

plt.loglog(nu_low, epsilon_low, linestyle='dotted', color='red', label=r'Auto-absorption: $\nu^{5/2}$')
plt.loglog(nu_mid, epsilon_mid, linestyle='solid', color='red', label=r'Optiquement mince: $\nu^{-\alpha}$')
plt.loglog(nu_high, epsilon_high, linestyle='dashed', color='red', label=r'Coupure: $\nu^{-\alpha} e^{-\nu/\nu_c}$')

# Ajout des lignes verticales pour les fréquences critiques
plt.axvline(nu_a, linestyle='--', color='b', label=r'$\nu_a$ (auto-absorption)')
plt.axvline(nu_c, linestyle='--', color='g', label=r'$\nu_c$ (coupure synchrotron)')

# Étiquettes et mise en page
plt.xlabel(r'Fréquence $\nu$ (Hz)')
plt.ylabel(r'Émissivité $\epsilon(\nu)$ (normalisée)')
# plt.title('Allure typique de l\'émissivité synchrotron par régime')
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Affichage du graphique
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes physiques
m_e = 9.109e-28  # masse de l'électron en g
c = 3.0e10  # vitesse de la lumière en cm/s
sigma_T = 6.65e-25  # section efficace de Thomson en cm^2
pi = np.pi

# Plage de valeurs pour gamma
gamma_values = np.logspace(2, 6, 100)  # Facteur de Lorentz de 100 à 1e6

# Champ magnétique fixé à 1 mG (converti en Tesla)
B_fixed = 1e-3  # 1 mG en Gauss

# Calcul du temps de refroidissement
t_s_gamma = (8 * pi * m_e * c) / (sigma_T * (B_fixed**2) * gamma_values)

# Conversion en années
t_s_gamma_years = t_s_gamma / (60 * 60 * 24 * 365.25)  # Conversion secondes → années

# Tracé de la courbe
plt.figure(figsize=(8, 5))
plt.loglog(gamma_values, t_s_gamma_years)

plt.xlabel(r'$\gamma$')
plt.ylabel(r'$t_s$ (années)')
# plt.title('Temps de refroidissement synchrotron en fonction de $\gamma$')
plt.grid(True, which="both", linestyle="--")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition des fréquences avec une gamme plus large vers les hautes énergies
nu = np.logspace(6, 19, 600)  # Fréquence de 10^6 à 10^30 Hz

# Paramètres des bosses spectrales
nu_syn_peak = 1e9  # Pic synchrotron à 10^9 Hz
gamma_e = 1e4  # Facteur de Lorentz des électrons
nu_ssc_peak = gamma_e**2 * nu_syn_peak  # Pic SSC basé sur gamma_e^2 * nu_syn_peak

# Intensités normalisées (loi de puissance avec coupure exponentielle)
L_syn = (nu / nu_syn_peak)**1.2 * np.exp(-nu / nu_syn_peak)
L_ssc = 0.1 * (nu / nu_ssc_peak)**1.2 * np.exp(-nu / nu_ssc_peak)

# Total du spectre
L_total = L_syn + L_ssc

# Suppression des valeurs inférieures à 10^-50
min_luminosity = 1e-30
L_total[L_total < min_luminosity] = min_luminosity

# Tracé du spectre avec une meilleure visibilité du total
plt.figure(figsize=(8, 5))
plt.loglog(nu, L_total, label="Spectre total", color="black", linewidth=2)  # Augmentation de l'épaisseur
plt.loglog(nu, L_syn, "--", label="Spectre synchrotron", color="blue", linewidth=1)
plt.loglog(nu, L_ssc, "--", label="Spectre SSC", color="red", linewidth=1)

# Labels et légende
plt.xlabel(r"Fréquence $\nu$ (Hz)")
plt.ylabel(r"Luminosité spectrale $L_{\nu}$ (normalisée)")
# plt.title("Spectre en double bosse (Synchrotron + SSC) avec $\gamma_e = 10^4$")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.ylim(min_luminosity, None)  # Fixe la limite inférieure à 10^-50
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Définition des paramètres
y = np.linspace(0, 1, 100)  # Distance dans le milieu
I_0 = 1.0  # Intensité initiale
j_y = 1.0  # Coefficient d'émission uniforme
tau_y = 3 * y  # Approximation d'une profondeur optique croissante

# Calcul des contributions
I_attenuation = I_0 * np.exp(-tau_y)  # Contribution de l'absorption
I_emission_integral = np.cumsum(j_y * np.exp(-tau_y)) * (y[1] - y[0])
I_total_corrected = I_attenuation + I_emission_integral  # Correction de l'intensité totale

# Tracé des courbes
plt.figure(figsize=(8, 5))
plt.plot(y, I_total_corrected, 'r-', label=r"Intensité totale")
plt.plot(y, I_attenuation, 'g--', label=r"Atténuation de l'intensité initiale")
plt.plot(y, I_emission_integral, 'b--', label=r"Cumul de l'émission et absorption locale")


# Labels et légendes

plt.xlabel("$s$")
plt.ylabel("Intensité $I$")
# plt.title("Propagation de l'intensité avec correction des contributions")
plt.legend()
plt.grid()
plt.show()
```
